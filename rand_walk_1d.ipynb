# random walk 1-d

## mean distance

```python
import random as rnd

Nrep = 100
Nstep = 21
sx = [0] * Nstep

# generate steps for random walk
for j in range(Nrep):
    x = [0] * Nstep
    for i in range(0, Nstep):
        d = 2*rnd.randint(0, 1) - 1
        x[i] = d
    for i in range(0, Nstep):
        sx[i] = sx[i] + x[i]

# mean distance for Nrep
for i in range(0, Nstep):
    sx[i] = sx[i] / Nrep

import matplotlib.pyplot as plt
plt.plot(sx)
plt.xlabel('steps')
plt.ylabel('average distance')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_random_walk(Nstep, Nrep):
    """Simulasi random walk 1D, return posisi pada setiap langkah."""
    positions = np.zeros((Nrep, Nstep))
    for r in range(Nrep):
        steps = np.random.choice([-1, 1], size=Nstep)
        positions[r] = np.cumsum(steps)
    return positions

# parameter simulasi
Nstep = 21
Nreps = [25, 50, 100, 200, 400]

mean_abs = {}

for Nrep in Nreps:
    pos = simulate_random_walk(Nstep, Nrep)
    mean_abs[Nrep] = np.mean(np.abs(pos), axis=0)

steps = np.arange(1, Nstep + 1)
theory_meanabs = np.sqrt(2 * steps / np.pi)   # teori E|x|  (2N/)

plt.figure(figsize=(8, 5))
for Nrep in Nreps:
    plt.plot(steps, mean_abs[Nrep], marker='o', label=f'Nrep={Nrep}')
plt.plot(steps, theory_meanabs, 'k--', label='theory (2N/)')
plt.xlabel('steps (N)')
plt.ylabel('mean distance |x|')
plt.title('Assignment 1  Mean Distance vs Steps')
plt.grid(True)
plt.legend()
plt.show()
```

## mean squared distance

```python
import random as rnd

Nrep = 10
Nstep = 121
sx2 = [0] * Nstep

# generate steps for random walk
for j in range(Nrep):
    pos = 0
    for i in range(0, Nstep):
        d = 2*rnd.randint(0, 1) - 1
        pos = pos + d
        sx2[i] = sx2[i] + pos * pos

# mean squared distance for Nrep
for i in range(0, Nstep):
    sx2[i] = sx2[i] / Nrep

import matplotlib.pyplot as plt
plt.plot(sx2)
plt.xlabel('steps')
plt.ylabel('mean squared distance')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_random_walk(Nstep, Nrep):
    """Simulasi random walk 1D, return posisi pada setiap langkah."""
    positions = np.zeros((Nrep, Nstep))
    for r in range(Nrep):
        steps = np.random.choice([-1, 1], size=Nstep)
        positions[r] = np.cumsum(steps)
    return positions

# parameter simulasi
Nstep = 121
Nreps = [25, 50, 100, 200, 400]

mean_sq = {}

for Nrep in Nreps:
    pos = simulate_random_walk(Nstep, Nrep)
    mean_sq[Nrep] = np.mean(pos ** 2, axis=0)

steps = np.arange(1, Nstep + 1)
theory_msd = steps   # teori E[x] = N

plt.figure(figsize=(8, 5))
for Nrep in Nreps:
    plt.plot(steps, mean_sq[Nrep], marker='o', label=f'Nrep={Nrep}')
plt.plot(steps, theory_msd, 'k--', label='theory x = N')
plt.xlabel('steps (N)')
plt.ylabel('mean squared distance x')
plt.title('Assignment 2  Mean Squared Distance vs Steps')
plt.grid(True)
plt.legend()
plt.show()
```
